````python
import numpy as np
from sentence_transformers import SentenceTransformer

# Initialize the SentenceTransformer model
model = SentenceTransformer('all-mpnet-base-v2')

def get_embedding(text):
    """
    Gets the embedding for the input text using Sentence Transformers.

    Args:
        text: The input text.

    Returns:
        The embedding vector for the input text.
    """
    return model.encode(text)

def compare_text_with_external_data(text, external_data_source):
    """
    Compares the input text with text from external data sources using embeddings.

    Args:
        text: The input text to compare.
        external_data_source: A list of text strings from the external data source.

    Returns:
        A list of tuples, where each tuple contains:
            - The text from the external data source.
            - The similarity score between the input text and the external text.
    """
    input_embedding = get_embedding(text)
    similarity_scores = []

    for external_text in external_data_source:
        external_embedding = get_embedding(external_text)
        similarity = np.dot(input_embedding, external_embedding) / (np.linalg.norm(input_embedding) * np.linalg.norm(external_embedding))
        similarity_scores.append((external_text, similarity))

    similarity_scores.sort(key=lambda x: x[1], reverse=True) # Sort by similarity
    return similarity_scores

# Example usage:
text = "This is a sample text to be compared"  # Your input text

# Example external data source (replace with your actual data)
external_data_source = [
    "This is a similar text to the input text.",
    "This is an unrelated text.",
    "Another text somewhat similar to the input.",
]

# Compare the text
results = compare_text_with_external_data(text, external_data_source)

# Print the results
print("Comparison Results:")
for external_text, similarity in results:
    print(f"External Text: {external_text}")
    print(f"Similarity Score: {similarity}")
    print("-" * 20)

# Further process for better recommendations
# based on the similarity scores
# For example, you might filter for similarity above a threshold or use the scores to rank the recommendations.
````

````python
import numpy as np
from sentence_transformers import SentenceTransformer


# Initialize the SentenceTransformer model
model = SentenceTransformer('all-mpnet-base-v2')

def get_embedding(text):
    """
    Gets the embedding for the input text using Sentence Transformers.
    """
    return model.encode(text)

def compare_text_with_gemini(text, api_key, num_candidates=3):
    """
    Compares the input text with Gemini-generated texts using embeddings.

    Args:
        text: The input text to compare.
        api_key: Your Gemini API key.
        num_candidates: Number of sentences Gemini should generate.

    Returns:
        A sorted list of tuples (generated_text, similarity_score).
    """
    input_embedding = get_embedding(text)
    similarity_scores = []

    # Configure Gemini API key
    genai.configure(api_key=api_key)

    # Initialize the Gemini model
    model = genai.GenerativeModel(model_name="gemini-2.0")

    # Generate similar sentences using Gemini
    prompt = f"Generate {num_candidates} unique and diverse sentences similar in meaning to: '{text}'"
    response = model.generate_content(prompt)

    # Parse and compare
    if isinstance(response.text, str):
        sentences = response.text.strip().split('\n')
        for sentence in sentences:
            sentence = sentence.strip('- ').strip()  # Remove leading dashes or bullets
            if sentence:
                external_embedding = get_embedding(sentence)
                similarity = np.dot(input_embedding, external_embedding) / (
                    np.linalg.norm(input_embedding) * np.linalg.norm(external_embedding)
                )
                similarity_scores.append((sentence, similarity))

    # Sort by similarity score
    similarity_scores.sort(key=lambda x: x[1], reverse=True)
    return similarity_scores

````
